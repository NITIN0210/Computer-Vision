```python
for i in range(0,270,30):
    gray = cv2.imread(r"E:\u\sem1\cv\aat3\question1\imagesFromVideo\frame%d.jpg"%i, 0)
    patch= cv2.imread(r"E:\u\sem1\cv\aat3\question1\cropedImage\frame30.jpg", 0)
    ssd = cv2.matchTemplate(gray, patch, cv2.TM_SQDIFF_NORMED) 
    point = np.where(ssd == ssd.min())
    y = point[0][0]
    x = point[1][0]
    w = len(patch[0])
    h = len(patch)
    plt.figure(figsize = (20,2))
    cv2.rectangle(gray, (x, y), (x+w, y+h), (255, 2, 1), 2)
    plt.axis('off')
    plt.imshow(gray,cmap='gray', vmin=0, vmax=255)
    plt.show()
```
